```python
from tensorflow.keras import Input, Model
```

```python
inputs = Input(shape=(n_features,))
x = Dense(32, activation='relu')(inputs)
x = Dense(16, activation='relu')(x)
x = Dense(8, activation='relu')(x)
outputs = Dense(1, activation='sigmoid')(x)

functional_model = Model(inputs=inputs, outputs=outputs)
```

```python
functional_model.summary()
```

```python
functional_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
functional_history = functional_model.fit(X_train, y_train, epochs=200, batch_size=16, verbose=1)
```

```python
loss, acc = functional_model.evaluate(X_test, y_test, verbose=1)
print('Functional Model Test Accuracy: %.3f' % acc)
```
